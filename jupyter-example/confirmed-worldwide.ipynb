# COVID19DB PROJECT (www.covid19db.org): covid19 - confirmed map


This notebook demonstrates how to connect to COVID19db Postgres database and how to generate simple visualisaions.

```python
import psycopg2
import pandas as pd
from pandas.plotting import register_matplotlib_converters 
register_matplotlib_converters()
import matplotlib.pyplot as plt

# Connect to covid19db.org
conn = psycopg2.connect(
    host='covid19db.org',
    port=5432,
    dbname='covid19',
    user='covid19',
    password='covid19'
)
cur = conn.cursor()
cur.execute("""SELECT i.* FROM
                (SELECT max(date) AS date, countrycode 
                FROM epidemiology WHERE source='WRD_ECDC' AND countrycode<>''
                GROUP BY countrycode ) AS recent_data
            INNER JOIN epidemiology AS i
            ON i.date=recent_data.date 
            AND i.countrycode = recent_data.countrycode
            AND i.adm_area_1 IS NULL
            AND i.source='WRD_ECDC'
            ORDER by i.confirmed DESC
            """)
covid19_data = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])
   
```

```python
import json
import geopandas as gpd

df = gpd.read_file('data/countries_geometries.geojson')
df.head()
```

```python
covid19_geo_data = df.merge(
    covid19_data[['countrycode','confirmed']], 
    how = 'left', left_on = 'ADM0_A3', right_on = 'countrycode'
)
covid19_geo_data.fillna({'confirmed':0}, inplace=True)
```

```python
import geoplot as gplt
import geoplot.crs as gcrs
import mapclassify as mc

schema = mc.Quantiles(covid19_geo_data['confirmed'], k=10)
gplt.choropleth(
    covid19_geo_data,  
    hue='confirmed',     
    edgecolor='white', 
    linewidth=1,
    cmap='Reds',
    scheme=schema,
    legend=True, 
    legend_kwargs={'bbox_to_anchor': (1, 0.9), 'frameon': False},
)
```

### Interactive map with folum:


```python
import folium
import numpy as np

world_map = folium.Map(location=[11,0], tiles="cartodbpositron", zoom_start=2, max_zoom = 6, min_zoom = 2)

folium.Choropleth(
    geo_data='data/countries_geometries.geojson',
     data=covid19_data,
     columns=['countrycode', 'confirmed'],
     key_on='feature.properties.ADM0_A3',
     fill_color='YlOrRd', 
     fill_opacity=0.7, 
     line_opacity=0.2,
     highlight=True,
     legend_name='COVID19 infected',
     nan_fill_color='black',
     nan_fill_opacity=0.4,
).add_to(world_map)
```
