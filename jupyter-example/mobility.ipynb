# COVID19DB PROJECT (www.covid19db.org): covid19 - Mobility data


This notebook demonstrates how to connect to COVID19db Postgres database and how to access Mobility data.

```python
import psycopg2
import pandas as pd
from pandas.plotting import register_matplotlib_converters 

# Connect to covid19db.org
conn = psycopg2.connect(
    host='covid19db.org',
    port=5432,
    dbname='covid19',
    user='covid19',
    password='covid19'
)
cur = conn.cursor()
cur.execute("""SELECT * FROM mobility
               WHERE source = 'GOOGLE_MOBILITY' AND
                     countrycode='GBR' AND 
                     adm_area_1 IS NULL AND
                     adm_area_2 IS NULL
               ORDER BY date
            """)

mobility_data = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])
mobility_data.head()
```
