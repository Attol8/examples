# OxCOVID19 (https://covid19.eng.ox.ac.uk/) - Confirmed and dead for selected countries

This notebook demonstrates how to connect to COVID19db Postgres database and how to generate simple visualisations.

```python
import psycopg2
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
%matplotlib inline
import matplotlib.pyplot as plt

# Connect to covid19db.org
conn = psycopg2.connect(
    host='covid19db.org',
    port=5432,
    dbname='covid19',
    user='covid19',
    password='covid19'
)
cur = conn.cursor()
cur.execute("SELECT * FROM epidemiology;")
df = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])


# Choose countries and perform initial preprocesing
my_countries = ['ITA', 'ESP', 'GBR', 'USA']

df_who = []
for countrycode in my_countries:
    df_who.append(df[(df.source=='WRD_ECDC') & (df.countrycode==countrycode) & df['adm_area_1'].isnull()].sort_values(by='date'))
```

## Confirmed and Dead for selected countries

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
for j in range(len(my_countries)):
    plt.plot(df_who[j].date, df_who[j].confirmed, linewidth=4.0, label=my_countries[j])
plt.title('Confirmed')
plt.grid()
plt.legend();

plt.subplot(1,2,2)
for j in range(len(my_countries)):
    plt.plot(df_who[j].date, df_who[j].dead, '-o', linewidth=3.0, label=my_countries[j])
plt.title('Dead')
plt.grid()
plt.legend();
```

## Deaths total and per day (starting with 10 deaths)

```python
plt.figure(figsize=(15,5))
start_with_deaths = 10;

plt.subplot(1,2,1)
for j in range(len(my_countries)):
    plt.plot(range(len(df_who[j].dead[df_who[j].dead>=start_with_deaths])),df_who[j].dead[df_who[j].dead>=start_with_deaths],'-o', linewidth=3.0,label=my_countries[j])
plt.legend();
plt.title('Total deaths (starting with 10 deaths)');

plt.subplot(1,2,2)
for j in range(len(my_countries)):
    plt.plot(range(len(df_who[j].dead[df_who[j].dead>=start_with_deaths])),df_who[j].dead[df_who[j].dead>=start_with_deaths].diff(),'-o', linewidth=3.0,label=my_countries[j])
plt.legend();
plt.title('Deaths per day (starting with 10 deaths)');
```
